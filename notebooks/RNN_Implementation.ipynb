# Give the network a sequence of words and train it to predict the next word.

Number of words is left as a parameter; we’ll use 50 for the examples shown here which means we give our network 50 words and train it to predict the 51st. Other ways of training the network would be to have it predict the next word at each point in the sequence — make a prediction for each input word rather than once for the entire sequence — or train the model using individual characters. The implementation used here is not necessarily optimal — there is no accepted best solution — but it works well!



```python
features=[]
labels=[]
train_ex=50

for i in sequences:
    for j in range(0,len(i)):
        features.append(i[j:train_ex+j])
        
        #labels.append(i[train_ex+j+1:])
    #print(len(i))
```
